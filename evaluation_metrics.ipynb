# **BLEU SCORE** **Calculation**

```python
import numpy as np
import openpyxl
from evaluate import load
```

```python
bleu_scorer = load("bleu")
```

```python
import openpyxl
from nltk.translate.bleu_score import sentence_bleu
from nltk.translate.nist_score import sentence_nist

def custom_tokenizer(text, ignore_words):
    tokens = text.split()
    filtered_tokens = [token for token in tokens if token.lower() not in ignore_words]
    return filtered_tokens

ignore_words = {'[PATCH]', 'Bugsplainer', 'Bugsplainer 220M'}

workbook = openpyxl.load_workbook('/content/drive/MyDrive/Colab Notebooks/pysecdb-security-50.xlsx')

bleu_scores= []
for worksheet_name in ['security', 'non-security-BP']:
    worksheet = workbook[worksheet_name]

    worksheet.cell(row=1, column=worksheet.max_column + 1, value="BLEU Score")

    for row in range(2, worksheet.max_row + 1):
        commit_message = worksheet.cell(row=row, column=5).value
        bugsplainer_text = worksheet.cell(row=row, column=6).value
        if commit_message is not None and bugsplainer_text is not None:
            commit_message_tokens = custom_tokenizer(commit_message, ignore_words)
            bugsplainer_tokens = custom_tokenizer(bugsplainer_text, ignore_words)

            bleu_score = sentence_bleu([commit_message_tokens], bugsplainer_tokens, weights=(1, 0, 0, 0))
            bleu_score_percentage = bleu_score * 100
            bleu_scores.append(bleu_score_percentage)

            worksheet.cell(row=row, column=worksheet.max_column, value=bleu_score_percentage)

workbook.save('/content/drive/MyDrive/Colab Notebooks/pysecdb-bleu-score-1.xlsx')
```

```python
print(bleu_scores)
```

# **ROUGE SCORE** and **BERT SCORE** **Calculation**

```python
import numpy as np
import openpyxl
from evaluate import load
```

```python
rouge_scorer = load("rouge")
bert_scorer = load("bertscore")
```

```python
def custom_tokenizer(text, ignore_words):
    tokens = text.split()
    filtered_tokens = [token for token in tokens if token.lower() not in ignore_words]
    return filtered_tokens

ignore_words = {'[PATCH]', 'Bugsplainer', 'Bugsplainer 220M'}

workbook = openpyxl.load_workbook('/content/drive/MyDrive/Colab Notebooks/pysecdb-security-50.xlsx')

rouge_scores= []
for worksheet_name in ['security', 'non-security-BP']:
    worksheet = workbook[worksheet_name]

    worksheet.cell(row=1, column=worksheet.max_column + 1, value="ROUGE Score1")
    worksheet.cell(row=1, column=worksheet.max_column + 1, value="ROUGE Score2")
    worksheet.cell(row=1, column=worksheet.max_column + 1, value="ROUGE ScoreL")
    worksheet.cell(row=1, column=worksheet.max_column + 1, value="ROUGE ScoreLSum")


    for row in range(2, worksheet.max_row + 1):
        commit_message = worksheet.cell(row=row, column=5).value
        bugsplainer_text = worksheet.cell(row=row, column=6).value
        if commit_message is not None and bugsplainer_text is not None:
            commit_message_tokens = custom_tokenizer(commit_message, ignore_words)
            bugsplainer_tokens = custom_tokenizer(bugsplainer_text, ignore_words)

            rouge_score = rouge_scorer.compute(predictions=[bugsplainer_text], references=[commit_message])
            for key, score in rouge_score.items():
              rouge_score[key] = score * 100
            rouge_scores.append(rouge_score)

            worksheet.cell(row=row, column=worksheet.max_column - 3 , value=rouge_score['rouge1'])
            worksheet.cell(row=row, column=worksheet.max_column - 2, value=rouge_score['rouge2'])
            worksheet.cell(row=row, column=worksheet.max_column - 1, value=rouge_score['rougeL'])
            worksheet.cell(row=row, column=worksheet.max_column, value=rouge_score['rougeLsum'])

workbook.save('/content/drive/MyDrive/Colab Notebooks/pysecdb-rouge-score-1.xlsx')
```

## **BERT SCORE Calculation**

```python
def custom_tokenizer(text, ignore_words):
    tokens = text.split()
    filtered_tokens = [token for token in tokens if token.lower() not in ignore_words]
    return filtered_tokens

ignore_words = {'[PATCH]', 'Bugsplainer', 'Bugsplainer 220M'}

workbook = openpyxl.load_workbook('/content/drive/MyDrive/Colab Notebooks/pysecdb-security-50.xlsx')

bert_scores= []
for worksheet_name in ['security', 'non-security-BP']:
    worksheet = workbook[worksheet_name]

    worksheet.cell(row=1, column=worksheet.max_column + 1, value="BERT Score - Precision")
    worksheet.cell(row=1, column=worksheet.max_column + 1, value="BERT Score - Recall")
    worksheet.cell(row=1, column=worksheet.max_column + 1, value="BERT Score - F1")

    for row in range(2, worksheet.max_row + 1):
        commit_message = worksheet.cell(row=row, column=5).value
        bugsplainer_text = worksheet.cell(row=row, column=6).value
        if commit_message is not None and bugsplainer_text is not None:
            commit_message_tokens = custom_tokenizer(commit_message, ignore_words)
            bugsplainer_tokens = custom_tokenizer(bugsplainer_text, ignore_words)

            bert_score = bert_scorer.compute(predictions=[bugsplainer_text], references=[commit_message], lang='en')
            for key, score in bert_score.items():
              if key!= 'hashcode':
                bert_score[key] = score[0] * 100
            bert_scores.append(bert_score)

            worksheet.cell(row=row, column=worksheet.max_column - 2 , value=bert_score['precision'])
            worksheet.cell(row=row, column=worksheet.max_column - 1, value=bert_score['recall'])
            worksheet.cell(row=row, column=worksheet.max_column, value=bert_score['f1'])

workbook.save('/content/drive/MyDrive/Colab Notebooks/pysecdb-bert-score-1.xlsx')
```
